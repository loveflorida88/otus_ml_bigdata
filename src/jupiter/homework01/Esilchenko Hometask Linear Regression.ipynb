```scala
import $ivy.`com.github.tototoshi::scala-csv:1.3.6`
import com.github.tototoshi.csv._
import java.io._
import $ivy.`org.scalanlp::breeze:1.0`
import breeze.linalg._
import breeze.numerics._
import breeze.stats.regression._
import scala.util.{Try, Success, Failure}
```

```scala
def load_dataset(path: String): DenseMatrix[Double]  = {
    val csv_reader = CSVReader.open(new File(path))
    val dataset = csv_reader.all().map(x=> x.toArray)
    val double_dataset = dataset.map(_.map(x => Try(x.toDouble).getOrElse(1.0)))
    DenseMatrix(double_dataset:_*)
}
```

```scala
def get_features(dataset: DenseMatrix[Double]): DenseMatrix[Double]  = {
    val ones = DenseVector.ones[Double](dataset(::,0).length)
    DenseMatrix.horzcat( ones.asDenseMatrix.t, dataset(::, 0 to 26))
}
```

```scala
def rmse(predictions: DenseVector[Double], targets: DenseVector[Double]): Double = {
    sqrt(sum((predictions - targets)*:*(predictions - targets))/predictions.length)
}
```

```scala
val facebookCommentFilePath: String = "/home/jovyan/data/Dataset.csv"
```

```scala
val facebookCommentDataset = load_dataset(facebookCommentFilePath)
```

```scala
val features = get_features(facebookCommentDataset)
```

```scala
val targets = facebookCommentDataset(::, -1)
```

```scala
val linear_regression = features*leastSquares(features, targets).coefficients
```

```scala
val lasso_regression = features*lasso(features, targets, 0.02).coefficients
```

```scala
val simple_regression =  inv(features.t * features) * features.t * targets 
```

```scala
val result_simple_regression = features * simple_regression
```

```scala
println("RMSE простой регрессии breeze:" + rmse(linear_regression, targets))
println("RMSE lasso:" + rmse(lasso_regression, targets))
println("RMSE своя простая регрессия: " + rmse(result_simple_regression, targets))
```
